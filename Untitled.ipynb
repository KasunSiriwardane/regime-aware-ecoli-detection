```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score
import pickle

# --- CONFIGURATION ---
MODEL_PATH = '../models/moe_model_final.pkl'
DATA_PATH = '../data/processed/Chattahoochee_Expert_Ready.csv'
EPA_THRESHOLD_LOG = np.log10(235 + 1)

# 1. Load Data
df = pd.read_csv(DATA_PATH)
df['Date'] = pd.to_datetime(df['Date'])
target_col = 'Log_Ecoli'

# Define Features (Same as Notebook 3)
cols_to_drop = ['Date', 'Log_Ecoli', 'Ecoli_CFU', 'Remark', 'site_no', 'Target_Unsafe', 'Cond_Filled', 'Regime_Storm']
features = [c for c in df.columns if c not in cols_to_drop and not c.endswith('_cd')]

# --- CRITICAL FIX: FILL NANs FOR SKLEARN ---
X = df[features].fillna(0) # Logistic Regression needs this
y = df[target_col]
binary_y = (y > EPA_THRESHOLD_LOG).astype(int)

# Vault Split
split_idx = int(len(df) * 0.85)
X_train = X.iloc[:split_idx]
y_train = binary_y.iloc[:split_idx]
X_vault = X.iloc[split_idx:]
y_vault = binary_y.iloc[split_idx:]

print(f"Vault Size: {len(X_vault)}")

# --- BASELINE 1: THE "DUMB" HEURISTIC ---
# Rule: If it rained > 0.5 inches in last 3 days -> ALARM
rain_threshold = 0.5
# Ensure Rain_3Day_Sum exists
if 'Rain_3Day_Sum' in X_vault.columns:
    heuristic_preds = (X_vault['Rain_3Day_Sum'] > rain_threshold).astype(int)
    h_rec = recall_score(y_vault, heuristic_preds)
    h_prec = precision_score(y_vault, heuristic_preds, zero_division=0)
    h_f1 = f1_score(y_vault, heuristic_preds, zero_division=0)

    print("\n--- BASELINE 1: RAIN HEURISTIC (>0.5 inch) ---")
    print(f"Recall: {h_rec:.2%}")
    print(f"Precision: {h_prec:.2%}")
    print(f"F1: {h_f1:.4f}")
else:
    print("Skipping Heuristic (Feature Missing)")
    h_rec, h_prec, h_f1 = 0, 0, 0

# --- BASELINE 2: LOGISTIC REGRESSION ---
print("\n--- BASELINE 2: LOGISTIC REGRESSION ---")
lr = LogisticRegression(max_iter=5000, class_weight='balanced') 
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_vault)

l_rec = recall_score(y_vault, lr_preds)
l_prec = precision_score(y_vault, lr_preds, zero_division=0)
l_f1 = f1_score(y_vault, lr_preds, zero_division=0)

print(f"Recall: {l_rec:.2%}")
print(f"Precision: {l_prec:.2%}")
print(f"F1: {l_f1:.4f}")

# --- YOUR MODEL (MoE) RESULTS ---
# Hardcoded from your Notebook 3 output
moe_rec = 0.8302
moe_prec = 0.6286
moe_f1 = 0.7154

# --- COMPARISON CHART ---
results_df = pd.DataFrame({
    'Model': ['Rain Heuristic', 'Logistic Regression', 'MoE (Ours)'],
    'F1 Score': [h_f1, l_f1, moe_f1],
    'Recall': [h_rec, l_rec, moe_rec],
    'Precision': [h_prec, l_prec, moe_prec]
})

print("\n--- FINAL BENCHMARK ---")
print(results_df)

# Plot F1
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='F1 Score', data=results_df, palette='viridis')
plt.title("Performance Benchmark: Does Complexity Pay Off?", fontsize=14)
plt.ylabel("F1 Score")
plt.ylim(0, 1.0)
for index, row in results_df.iterrows():
    plt.text(index, row['F1 Score'] + 0.02, f"{row['F1 Score']:.3f}", color='black', ha="center", fontsize=12, fontweight='bold')
plt.show()

# --- FORENSIC VISUAL: STORM EVENT CASE STUDY ---
# Find max rain event in Vault
try:
    storm_idx = X_vault['Rain_3Day_Sum'].idxmax()
    storm_date = df.loc[storm_idx, 'Date']
    start_window = storm_date - pd.Timedelta(days=10)
    end_window = storm_date + pd.Timedelta(days=10)

    subset = df[(df['Date'] >= start_window) & (df['Date'] <= end_window)].copy()
    
    # Load MOE Model
    with open(MODEL_PATH, 'rb') as f:
        moe_model = pickle.load(f)
    
    # Predict
    # We need to recreate the input exactly for the model
    # Note: moe_model.predict needs the regime column from the dataframe context if we use the class method directly,
    # OR we need to manually route.
    # Let's try the class method first. It expects 'Regime_Storm' inside X or passed separately.
    # Our Class definition in NB3 took (X, regime).
    
    subset_X = subset[features]
    subset_regime = subset['Regime_Storm']
    
    subset_preds = moe_model.predict(subset_X, subset_regime)
    
    # Plot
    plt.figure(figsize=(12, 6))
    
    # 1. True Bacteria
    plt.plot(subset['Date'], subset[target_col], 'ko-', label='True Bacteria (Log)', markersize=6)
    
    # 2. Prediction
    plt.plot(subset['Date'], subset_preds, 'r--', linewidth=2, label='MoE Risk Prediction (90%)')
    
    # 3. Limit
    plt.axhline(EPA_THRESHOLD_LOG, color='green', label='Safety Limit', linestyle='-.')
    
    plt.ylabel("Log10 E. coli")
    plt.title(f"Forensic Analysis: Model Response to Storm Event ({storm_date.date()})", fontsize=14)
    plt.legend(loc='upper left')
    
    # 4. Rain on secondary axis
    ax2 = plt.gca().twinx()
    ax2.bar(subset['Date'], subset['Rain_inches'], color='blue', alpha=0.2, label='Rain')
    ax2.set_ylabel("Rain (inches)", color='blue')
    
    plt.grid(True, alpha=0.3)
    plt.show()

except Exception as e:
    print(f"Could not plot time series: {e}")
```

```python
# --- RE-DEFINE CLASS FOR PICKLE LOADING ---
# (Python needs this definition to reconstruct the object)
from sklearn.base import BaseEstimator, ClassifierMixin
import xgboost as xgb

class MixtureOfExperts(BaseEstimator, ClassifierMixin):
    def __init__(self, alpha=0.90):
        self.alpha = alpha
        self.expert_storm = xgb.XGBRegressor()
        self.expert_calm = xgb.XGBRegressor()
        self.correction_factor = 0.0
        
    def _get_storm_mask(self, X):
        return (X['Regime_Storm'] == 1) & (X['Days_Since_Rain'] <= 1)

    def predict_raw(self, X):
        is_storm = self._get_storm_mask(X)
        preds = np.zeros(len(X))
        if is_storm.sum() > 0:
            preds[is_storm] = self.expert_storm.predict(X[is_storm])
        if (~is_storm).sum() > 0:
            preds[~is_storm] = self.expert_calm.predict(X[~is_storm])
        return preds

    def predict(self, X, regime=None): # Added regime arg flexibility
        # If regime is not passed explicitly, assume it's in X
        return self.predict_raw(X) + self.correction_factor

# --- FORENSIC VISUAL: STORM EVENT CASE STUDY ---
try:
    print("Loading Model for Visuals...")
    with open(MODEL_PATH, 'rb') as f:
        moe_model = pickle.load(f)
    
    # Find max rain event in Vault
    storm_idx = X_vault['Rain_3Day_Sum'].idxmax()
    storm_date = df.loc[storm_idx, 'Date']
    
    # Define Window (10 days before/after)
    start_window = storm_date - pd.Timedelta(days=10)
    end_window = storm_date + pd.Timedelta(days=10)
    subset = df[(df['Date'] >= start_window) & (df['Date'] <= end_window)].copy()
    
    # Predict
    # Ensure subset has all necessary columns for the model
    subset_preds = moe_model.predict(subset)
    
    # Plot
    plt.figure(figsize=(12, 6))
    
    # 1. True Bacteria
    plt.plot(subset['Date'], subset[target_col], 'ko-', label='True Bacteria (Log)', markersize=6)
    
    # 2. Prediction
    plt.plot(subset['Date'], subset_preds, 'r--', linewidth=2, label='MoE Risk Prediction (90%)')
    
    # 3. Limit
    plt.axhline(EPA_THRESHOLD_LOG, color='green', label='Safety Limit', linestyle='-.')
    
    plt.ylabel("Log10 E. coli")
    plt.title(f"Forensic Analysis: Model Response to Storm Event ({storm_date.date()})", fontsize=14)
    plt.legend(loc='upper left')
    
    # 4. Rain on secondary axis
    ax2 = plt.gca().twinx()
    ax2.bar(subset['Date'], subset['Rain_inches'], color='blue', alpha=0.2, label='Rain')
    ax2.set_ylabel("Rain (inches)", color='blue')
    
    plt.grid(True, alpha=0.3)
    plt.show()
    
except Exception as e:
    print(f"Could not plot time series: {e}")
```
